```python
import pandas as pd

# Actual May data
actual = pd.read_excel(
    "../data/May 7 Data.xlsx"
)

# Top-100 predicted call list
predicted = pd.read_csv(
    "../data/may 7 top 100.csv"
)

print("Actual:", actual.shape)
print("Predicted:", predicted.shape)
```

```python
actual["createdAt"] = pd.to_datetime(
    actual["createdAt"],
    dayfirst=True,
    errors="coerce"
)

may7 = actual[
    actual["createdAt"].dt.date
    ==
    pd.to_datetime("2026-05-07").date()
]

# Remove cancelled orders
may7 = may7[
    may7["orderStatus"] != "Cancelled"
]

actual_retailers = set(
    may7["customerId"]
    .astype(str)
    .unique()
)

print(
    "Actual retailers who ordered:",
    len(actual_retailers)
)
```

```python
print(predicted.columns.tolist())
```

```python
predicted_retailers = set(
    predicted["Retailer ID"]
    .astype(str)
    .unique()
)

print(
    "Predicted retailers:",
    len(predicted_retailers)
)
```

```python
matched = predicted_retailers.intersection(
    actual_retailers
)

print(
    "Matched retailers:",
    len(matched)
)
```

```python
precision_at_100 = (
    len(matched)
    /
    len(predicted_retailers)
)

recall_at_100 = (
    len(matched)
    /
    len(actual_retailers)
)

print(
    f"Precision@100 = {precision_at_100:.2%}"
)

print(
    f"Recall@100 = {recall_at_100:.2%}"
)
```
